Do the following operations on ‘cars_data’ dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 # 1. Read the dataset to the python environment

```python
rd=pd.read_csv(r"C:\Users\USER\Documents\DSA notes\cars_data.csv")
rd
```

# 2.Check for null values present in the dataset

```python
rd.isnull().sum()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(12,15))
sp=sns.countplot(x=rd['Buyer Gender'],data=rd)
sns.set_theme(style='whitegrid',palette='colorblind',font_scale=2)
for p in sp.patches:
    sp.annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9),
                   textcoords = 'offset points',size=20)
plt.title('Female vs Male Buyers',size=30)
plt.xticks(fontsize=20)
```

Insight:

Number of Female buyers participated in car sales is 2.1% higher than number of Males.
